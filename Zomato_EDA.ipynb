# Top 10 Cuisines

```python
final_df['Cuisines'] = final_df['Cuisines'].astype(str).fillna('')

# Split the Cuisines into a list of cuisines
cuisines_split = final_df['Cuisines'].str.split(', ')

# Flatten the list of lists and create a single Series
cuisines_flat = pd.Series([cuisine for sublist in cuisines_split for cuisine in sublist if isinstance(cuisine, str)])

# Count the occurrences of each cuisine
top_ten_cuisines = cuisines_flat.value_counts().head(10)

# Create a DataFrame from the top ten cuisines
top_ten_cuisines_df = top_ten_cuisines.reset_index()
top_ten_cuisines_df.columns = ['Cuisine', 'Count']

# Plot the top ten cuisines
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Cuisine', data=top_ten_cuisines_df)
plt.title('Top 10 Cuisines')
plt.xlabel('Count')
plt.ylabel('Cuisine')
plt.show()
```
